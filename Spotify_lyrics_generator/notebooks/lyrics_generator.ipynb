```python
# We will now create a RNN as our lyrics generation model (actually, it is a language model!)
# FIRST: Create sequences for our model to take as a seed and to learn:
total_text_as_list = []

def extract_text(lyrics_data):
    global total_text_as_list
    total_text_as_list += [w for w in lyrics_data.split(' ') if w.strip() != '' or w == '\n']
    return total_text_as_list

lyrics_data['list of words'] = lyrics_data['lyrics']
lyrics_data['list of words'].apply(extract_text)
print('Total words: ', len(total_text_as_list))
print('Total unique words: ', len(set(total_text_as_list)))

words = sorted(set(total_text_as_list))
regex = re.compile(r"[',!@#$;:!*?%)(&^~]")
filtered_words = [i for i in words if not regex.match(i)]
#print(filtered_words)

filtered_total_text_as_list = [i for i in total_text_as_list if not regex.match(i)]
#print(filtered_total_text_as_list)

corpus = (" ".join(filtered_total_text_as_list)).strip()
corpus
```

```python
# create a character mapping index
chars = sorted(list(set(corpus)))
mapping = dict((c, i) for i, c in enumerate(chars))
print(sorted(list(set(corpus))))
print(mapping)

# vocabulary size
vocab = len(mapping)
print("Encoding vocabulary size: ",vocab)

def encode_seq(seq):
    sequences = list()
    for line in seq:
        # integer encode line
        encoded_seq = [mapping[char] for char in line]
        #print(encoded_seq)
        # store
        sequences.append(encoded_seq)
    return sequences
```

```python
def create_seqs(sequences, min_seq):
    start_text_sequences = []
    end_text_sequences = []
    # select sequence of tokens
    for i in range(min_seq, len(corpus)-min_seq):
        start_text_sequences.append(("".join(corpus[i - min_seq: i])))
        end_text_sequences.append(("".join(corpus[i: i+min_seq])))
    print(f"Number of start sequences generated: {len(start_text_sequences)}")
    print(f"Number of end sequences generated: {len(end_text_sequences)}")
    return start_text_sequences, end_text_sequences

min_seq = vocab
start_text_sequences, end_text_sequences = create_seqs(filtered_total_text_as_list,min_seq)
```

```python
encoded_start_text_sequences = array(encode_seq(start_text_sequences))
encoded_end_text_sequences = array(encode_seq(end_text_sequences))

print(start_text_sequences[0])
print(encoded_start_text_sequences[0])
```

```python
X = encoded_start_text_sequences
y = encoded_end_text_sequences

# one hot encode X and y
#X = array([to_categorical(x, num_classes=len(mapping)) for x in X])
y = array(to_categorical(encoded_end_text_sequences, num_classes=vocab))

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.01, random_state=20)
#X_train, X_test, y_train, y_test = train_test_split(sequences_start, sequences_end, test_size=0.05, random_state=50)
print('Train shape:', X_train.shape, 'Val shape:', X_test.shape)
print(X_train[6:10])
```

```python
print(X[0])
print(X.shape)
```

```python
def get_model(X):
    model = Sequential()
    model.add(Embedding(vocab, 50, input_length=vocab, trainable=True))
    model.add(LSTM(75, input_length=(X.shape[0], X.shape[1]), return_sequences=True))
    model.add(Dense(vocab, activation='softmax'))
    print(model.summary())
    model.compile(loss='categorical_crossentropy', metrics=['acc'], optimizer='adam')
    return model

model = get_model(X)
# fit the model
history = model.fit(X_train, y_train, epochs=150, verbose=2, validation_data=(X_test, y_test))
```

```python
loss = history.history['loss']
fig = plt.figure()
plt.plot(loss)
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show();
```

```python
# GENERATING LYRICS:
lyrics_data['list of phrases'] = lyrics_data['lyrics']

def split_phrases(string):
    b = string.split("\n")
    return b

#print(lyrics_data)

def create_seeds_from_phrases(lyrics_data):
    lyrics_data_local = lyrics_data
    seed_list = []
    lyrics_data_local['list of phrases'] = lyrics_data_local['list of phrases'].apply(split_phrases)
    preprocessed_list = list(lyrics_data_local['list of phrases'])
    for i in range(len(preprocessed_list)):
        for j in range(len(preprocessed_list[i])):
            seed_list.append(preprocessed_list[i][j])
    seed_list = [i for i in seed_list if i != ' ' and i != '  ' and i != '   ']
    return list(set(seed_list))

#seed_list = create_seeds_from_phrases(lyrics_data)
#seed_list
```

```python
def generate_lyrics(size_gen,phrases,seed="",model=model,mapping=mapping,seq_length=min_seq,lyrics_data=lyrics_data):
    num_classes = len(mapping)
    if seed == "":
        seed_list=[]
        seed_list = create_seeds_from_phrases(lyrics_data)
        seed_index = np.random.randint(len(seed_list))
        seed = seed_list[seed_index]
    txt = seed + '\n'
    print("SEED: ",seed)
    print("================================================================")
    print("GENERATED LYRICS: ")
    # generate a fixed number of characters and phrases
    for j in range(phrases):
        for i in range(size_gen):
            encoded = pad_sequences([[mapping[c] for c in txt]], maxlen=seq_length, truncating='pre')
            encoded = to_categorical(encoded, num_classes=num_classes)
            #print(encoded.shape)
            ypred = np.argmax(model.predict(encoded,verbose=0),axis=-1)[0][0]
            output_char = ''
            for char, index in mapping.items():
                if index == ypred:
                    output_char = char
                    break
            # append to input
            txt += output_char
        txt += '\n'
    return txt

print(generate_lyrics(50,10))
```
