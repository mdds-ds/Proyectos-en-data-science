```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score
```

```python
df_mod = df_mod[df_mod["gender"].isna() == False]
```

```python
df_mod["gender"].value_counts()
```

```python
df_mod = df_mod[df_mod["gender"] != "Other"]
df_mod["gender"].value_counts()
```

```python
def cambia_gen (x):
  if "Female" in x:
    return 2
  elif "Male" in x:
    return 1
  else: 
    return 0

df_mod["gender"] = df_mod["gender"].apply(cambia_gen)
```

```python
features = ['basesalary', 'yearsofexperience', 'yearsatcompany','Doctorate_Degree', 'Highschool', 'Some_College', 'Bachelors_Degree','Masters_Degree','gender','totalyearlycompensation', "location"]

y = df_mod["Race"].to_numpy()


X = np.array(df_mod[features])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10)

clf = RandomForestClassifier(n_estimators= 500)

clf.fit(X_train, y_train)


y_hat_test = clf.predict_proba(X_test)#[:, 1]
y_hat_train = clf.predict_proba(X_train)#[:, 1]

# evaluo el AUC
roc_test = roc_auc_score(y_test, y_hat_test, multi_class = "ovr")
roc_train = roc_auc_score(y_train, y_hat_train, multi_class = "ovr")
print("train: " + str(roc_train) + " test: " + str(roc_test))




feature_importance = clf.feature_importances_ 
print(feature_importance)
```
