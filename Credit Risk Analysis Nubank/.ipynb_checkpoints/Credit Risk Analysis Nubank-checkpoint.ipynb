```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split, cross_validate, GridSearchCV, cross_val_score
from imblearn.under_sampling import RandomUnderSampler
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, classification_report
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import StratifiedKFold
from xgboost import XGBClassifier
```

```python
nubank_data = pd.read_csv("acquisition_train.csv") #Real data from Nubank
nubank_data
```

```python
# Here we can see that several features have missing values
nubank_data.info()
```

```python
# Here, we have to delete target_default where it is null, as it is our target variable
nubank_data.dropna(subset=["target_default"], inplace=True)
nubank_data.info()
```

```python
# number of unique observations per column, useful for choosing which columns to drop
nubank_data.nunique().sort_values()
```

```python
# number of non defaulters vs defaulters (spoiler: we are dealing with an unbalanced dataset)
frequence = nubank_data["target_default"].value_counts()
frequence
```

```python
# As there are columns that provide personal info (such as email) not useful for a model, we'll drop these columns.
# We'll keep hashed columns as they will be useful for a model when we de-hash them.
nubank_data.drop(labels=["facebook_profile","state","zip","channel","email", "lat_lon", "shipping_zip_code",
                         "ids","profile_phone_number","external_data_provider_credit_checks_last_2_year"],
                axis=1, inplace=True)
```

```python
nubank_data.head()
```

```python
# I'll see other columns that may be interesting to analyze, I want to see which values they have to know if they
# would be useful for building a model:
nubank_data_aux = nubank_data[["reason","n_defaulted_loans","real_state","n_bankruptcies","marketing_channel",
                               "job_name","user_agent","profile_tags","application_time_applied"]]
nubank_data_aux
```

```python
# As there are columns provide personal info (such as email) or irrelevant info (such as application_time_applied), we'll drop these columns.
nubank_data.drop(labels=["user_agent","profile_tags","application_time_applied","reason","shipping_state","job_name","marketing_channel","external_data_provider_first_name"],
                axis=1, inplace=True)
# We will also drop some columns such as marketing_channel, reason and job_name for performance issues, 
# and we'll see later if we could have considered it relevant or not
nubank_data.head()
```

```python
nubank_data.columns
```

```python
# show descriptive statistics, we'll look for outliers and other weird values
nubank_data.describe()
```

```python
# count of "inf" values in "reported_income"
print("infs: ", np.isinf(nubank_data['reported_income']).sum())
# count of values = -999 in "external_data_provider_email_seen_before"
print(nubank_data.loc[nubank_data['external_data_provider_email_seen_before'] == -999, 'external_data_provider_email_seen_before'].value_counts())
```

```python
# replace "inf" values with "nan"
nubank_data['reported_income'] = nubank_data['reported_income'].replace(np.inf, np.nan)

# replace "-999" values with "nan"
nubank_data.loc[nubank_data['external_data_provider_email_seen_before'] == -999, 'external_data_provider_email_seen_before'] = np.nan

#We replace these values with nan so we can plot a histogram of numerical features:
nubank_data_numerical = nubank_data.select_dtypes(include=['float64'])
nubank_data_numerical
```

```python
# plot a histogram for each of the features above 

nrows = 6
ncols = 3

fig, ax = plt.subplots(nrows=nrows, ncols=ncols, figsize=(30, 25))

r = 0
c = 0

for i in nubank_data_numerical:
    sns.distplot(nubank_data_numerical[i], bins=20,kde=False, ax=ax[r][c])
    if c == ncols - 1:
        r += 1
        c = 0
    else:
        c += 1

plt.show()
```

```python
# Now we see that these features are skewed, an indicator that they need to be filled with the median values of each feature

# As there are categorical and numerical features, categorical ones will be filled with the mode (most common value)
# while the numericals will be filled with the median.

# In the specific cases of last_amount_borrowed, last_borrowed_in_months and n_issues we'll fill the missing values
# with zero, as it is reasonable to believe that not every client would have values assigned to these variables.

nubank_data_numerical = nubank_data_numerical.columns
nubank_data_categorical = nubank_data.select_dtypes(include=['object']).columns

# fill missing values for "last_amount_borrowed", "last_borrowed_in_months" and "n_issues"
nubank_data['last_amount_borrowed'].fillna(value=0, inplace=True)
nubank_data['last_borrowed_in_months'].fillna(value=0, inplace=True)
nubank_data['n_issues'].fillna(value=0, inplace=True)

# fill missing values for numerical variables
imputer = SimpleImputer(missing_values=np.nan, strategy='median')
imputer = imputer.fit(nubank_data.loc[:, nubank_data_numerical])
nubank_data.loc[:, nubank_data_numerical] = imputer.transform(nubank_data.loc[:, nubank_data_numerical])

# fill missing values for categorical variables
imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imputer = imputer.fit(nubank_data.loc[:, nubank_data_categorical])
nubank_data.loc[:, nubank_data_categorical] = imputer.transform(nubank_data.loc[:, nubank_data_categorical])
```

```python
# We check that there aren't null values
nubank_data.isnull().sum()
```

```python
# We finally have a clean dataset to work with.
# Next step is to convert categorical to numerical using an encoder, so we can use them as input to our model.
# For binary categoricals, LabelEncoder will be used, for multi-valued categoricals, we'll use get_dummies
# Once this is done, we can start implementing our models.

binary_categorical = nubank_data.nunique()[nubank_data.nunique() == 2].keys().tolist()
numerical = [col for col in nubank_data.select_dtypes(['int', 'float']).columns.tolist() if col not in binary_categorical]
multival_categorical = [col for col in nubank_data.select_dtypes(['object']).columns.tolist() if col not in binary_categorical]

nubank_data_encoded = nubank_data.copy()

# label encoding for the binary variables
for col in binary_categorical:
    nubank_data_encoded[col] = LabelEncoder().fit_transform(nubank_data_encoded[col])
#print("OK")
```

```python
# encoding with get_dummies for the categorical variables
nubank_data_encoded = pd.get_dummies(nubank_data_encoded, columns=multival_categorical)
#sparse=True may be added to prevent kernel from dying
```

```python
nubank_data_encoded.head()
```

```python
# We'll just build a XGBoost model to predict default:
# feature matrix
X = nubank_data_encoded.drop('target_default', axis=1)

# target vector
y = nubank_data_encoded['target_default']

X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=True, stratify=y)
```

```python
# This is an unbalanced dataset, as there are much more non defaulters than defaulters (we verified that previously)
# standardize numerical variables
scaler = StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)

# resample
rus = RandomUnderSampler()
X_train_rus, y_train_rus = rus.fit_resample(X_train, y_train)
```

```python
"""Precision will give us the proportion of positive identifications that were indeed correct. It can be defined as:
 

Recall will determine the proportion of real positives that were
correctly identified, and it can be defined as:

 

F1 Score is a metric that is useful when we need to seek a balance between precision and recall. The formula is defined as:
 

Since our objective is to minimize company loss, predicting the risk of client default, a good recall rate is desirable because we want to identify the maximum amount of clients that are indeed prone to stop paying their debts, thus, we are pursuing a small number of False Negatives.

Additionally, we also seek to minimize the number of False Positives because we don't want clients to be mistakenly identified as defaulters. Therefore, a good precision rate is also desirable.

However, there is always a tradeoff between precision and recall. For this article, we chose to give more emphasis to recall, using it as our evaluation metric.

We'll use Cross-Validation to get better results. Instead of simply splitting the data into a train and test set, the cross_validate method splits our training data into k number of Folds, making better use of the data. In our case, we'll perform 5-fold cross-validation, as we let the default k value."""
```

```python
# define the function val_model
def validate_model(X, y, clf, show=True):
    """
    Apply cross-validation on the training set.

    # Arguments
        X: DataFrame containing the independent variables.
        y: Series containing the target vector.
        clf: Scikit-learn estimator instance.
        
    # Returns
        float, mean value of the cross-validation scores.
    """
    
    X = np.array(X)
    y = np.array(y)

    pipeline = make_pipeline(StandardScaler(), clf)
    scores = cross_val_score(pipeline, X, y, scoring='recall')

    if show == True:
        print(f'Recall: {scores.mean()}, {scores.std()}')
    
    return scores.mean()
```

```python
#evaluate the models
xgb = XGBClassifier()

model = []
recall = []

model.append(xgb.__class__.__name__)
recall.append(validate_model(X_train_rus, y_train_rus, xgb, show=False))

pd.DataFrame(data=recall, index=model, columns=['Recall'])
```
